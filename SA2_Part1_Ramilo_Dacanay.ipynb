#### Task 8 
Rename month names to English.

```python
df
```

```python
months = df['month'].tolist()
months
```

```python
month_eng = ['January','February','March','April','May','June','July','August','September','October','November','December']
for month in months:
        if month == 'Janeiro':
            df['month'] = df['month'].replace(month, month_eng[0])
        elif month == 'Fevereiro':
            df['month'] = df['month'].replace(month, month_eng[1])
        elif month == 'Março':
            df['month'] = df['month'].replace(month, month_eng[2])
        elif month == 'Abril':
            df['month'] = df['month'].replace(month, month_eng[3])
        elif month == 'Maio':
            df['month'] = df['month'].replace(month, month_eng[4])
        elif month == 'Junho':
            df['month'] = df['month'].replace(month, month_eng[5])
        elif month == 'Julho':
            df['month'] = df['month'].replace(month, month_eng[6])
        elif month == 'Agosto':
            df['month'] = df['month'].replace(month, month_eng[7])
        elif month == 'Setembro':
            df['month'] = df['month'].replace(month, month_eng[8])
        elif month == 'Outubro':
            df['month'] = df['month'].replace(month, month_eng[9])
        elif month == 'Novembro':
            df['month'] = df['month'].replace(month, month_eng[10])
        elif month == 'Dezembro':
            df['month'] = df['month'].replace(month, month_eng[11])
df
```
