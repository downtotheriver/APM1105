```python
def ack (a,b):
    if a==0:
        return ack_a_isZero(a,b)
    elif a>0 and b==0:
        return ack_b_isZero(a,b)
    else:
        return ack_a_b_isNotZero(a,b)
def ack_a_isZero(c,d):
    print(f"({c},{d}) = {d+1}")
    return d+1
def ack_b_isZero(e,f):
    f=1
    print(f"({e},{f}) = ({e-1},1)")
    return ack(e-1,f)
def ack_a_b_isNotZero(g,h):
    print(f'({g},{h}) = ({g-1},({g},{h-1}))')
    return ack(g-1,ack(g,h-1))
def userInput(): 
    i=1
    while i!=0:
        num_a=int(input("Input first number: "))
        num_b=int(input("Input second number: "))
        if num_a>=0 and num_b>=0:
            print(ack(num_a, num_b))
            i-=1
        else:
            print("\nINVALID INPUT")
            print("\nTry again\n")
            continue
def choose():
    choice=str(input("Do you wish to try again? (y/n): "))
    match choice:
        case "y"|"Y":
            userInput()
            return True
        case "n"|"N":
            print("Thank you for Using Ackerman")
            return False

        case _:
            print("\nINVALID INPUT\n")
            choose()
#A(2,4) part:
m=2
n=4
print("Welcome To Ackerman Function (2,4)")
print(ack(m, n))

#User Input:
print("Welcome To Ackerman Function User Oriented")
print("Please provide a positive Integer")
userInput()
w=1
while w!=0:
    if bool(choose())==False:
        w-=1
        break
    else:
        choose()
        
```
