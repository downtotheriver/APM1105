```python
import time


# 1. Create an exponent calculator using both iterations and recursion, with rendering time.
print("1. Create an exponent calculator using both iterations and recursion, with rendering time.")
def iterative_exponent(base, exponent):
    if base == 0 and exponent <= 0:
        return f"{base:.0f} raised to {exponent} is undefined"
    result = 1
    if exponent == 0:
        return 1
    if exponent < 0:
        base = 1 / base
        exponent = -exponent
    while exponent > 0:
        result *= base
        exponent -= 1
    return result


def recursive_exponent(base, exponent, initial_base):
    if base == 0 and exponent <= 0:
        return f"{base:.0f} raised to {exponent} is undefined"
    if exponent == 0:
        return 1
    if exponent < 0:
        return 1 / recursive_exponent(initial_base, -exponent, initial_base)
    return base * recursive_exponent(base, exponent - 1, initial_base)


while True:
    try:
        base = float(input("Enter the value of the base: "))
        break
    except ValueError:
        print("Please enter a valid float number.")
while True:
    try:
        exponent = int(input("Enter the value of the exponent: "))
        break
    except ValueError:
        print("Please enter a valid integer number.")

initial_base = base

start_time = time.perf_counter()
result_1 = iterative_exponent(base, exponent)
end_time = time.perf_counter()
iterative_runtime = end_time - start_time

start_time = time.perf_counter()
result_2 = recursive_exponent(base, exponent, initial_base)
end_time = time.perf_counter()
recursive_runtime = end_time - start_time

print("Iterative result:", result_1)
print("Iterative rending time:", iterative_runtime, "seconds")
print("Recursive result:", result_2)
print("Recursive rending time:", iterative_runtime, "seconds")

if iterative_runtime < iterative_runtime:
    print("Iterative is faster")
else:
    print("Recursion is faster")


# 2. Iteratively and recursively calculate the sum of the first n powers of 2, along with the rendering time of each function using the time module.
print("\n2. Iteratively and recursively calculate the sum of the first n powers of 2, along with the rendering time of each function using the time module.")


def iterative_power(exponent, base = 2):
    result = 0
    while exponent >= 0:
        result += base ** exponent
        exponent -= 1
    return result


def recursive_power(exponent, base=2):
    if exponent == 0:
        return 1
    else:
        return base ** exponent + recursive_power(exponent - 1, base)


while True:
    try:
        exponent = int(input("Enter the value of the exponent: "))
        if exponent < 0:
            raise ValueError("Please enter a non-negative number.")
        break  # Exit the loop if the input is valid
    except ValueError as e:
        print(e)

start_time = time.perf_counter()
result_1 = iterative_power(exponent)
end_time = time.perf_counter()
iterative_runtime = end_time - start_time

start_time = time.perf_counter()
result_2 = recursive_power(exponent)
end_time = time.perf_counter()
recursive_runtime = end_time - start_time

print("Iterative result:", result_1)
print("Iterative runtime:", iterative_runtime, "seconds")
print("Recursive result:", result_2)
print("Recursive runtime:", recursive_runtime, "seconds")

if iterative_runtime < iterative_runtime:
    print("Iterative is faster")
else:
    print("Recursion is faster")
```
