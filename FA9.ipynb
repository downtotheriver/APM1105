```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.image as mpimg
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

```python
df = pd.read_csv('Spotify_Youtube.csv')
```

```python
df['Likes'] = df['Likes'].fillna(0)
df['Comments'] = df['Comments'].fillna(0)
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df['Album_type'].value_counts()
```

```python
album_type_count = df['Album_type'].value_counts()
album_type_count = album_type_count / album_type_count.sum() * 100
labels = album_type_count.index.tolist()
sizes = album_type_count.values.tolist()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=55)
plt.title('Album Types')
plt.legend(labels, loc='best')
plt.show()

print('''Remarks:

    Majority of the types of album is album with 72.3% followed by single type of albums with 24.1% and lastly, compilation with 3.6%
    
    This concludes that majority of the artists usually release a whole album, while a small population of artists prefers to release an album type of compilation.''')
```

```python
# Group the songs by artist and views - youtube, stream - spotify
artist_grouped = df.groupby('Artist')[['Views', 'Stream']].sum()

# Sort the artists by the sum of views and streams in descending order
artist_sorted = artist_grouped.sort_values(['Views', 'Stream'], ascending=False)

# Get the top 10 artists with the most number of views on YouTube and streams on Spotify
top_10 = artist_sorted.head(10)
top_10
```

```python
# Define a function to add the logo to the chart
def add_logo(ax, logo_path):
    logo = mpimg.imread(logo_path)
    logo_ax = inset_axes(ax, width="20%", height="20%", loc=1)
    logo_ax.imshow(logo)
    logo_ax.axis('off')

# Group the data by artist and calculate the total views and streams
grouped_data_yt = df.groupby('Artist').agg({'Views': 'sum'}).sort_values(by=['Views'], ascending=False)[:10]
grouped_data_sp = df.groupby('Artist').agg({'Stream': 'sum'}).sort_values(by=['Stream'], ascending=False)[:10]

# Create a figure with 2 subplots side by side
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Create a bar chart for top 10 artists on YouTube
axs[0].bar(grouped_data_yt.index, grouped_data_yt['Views'], color='red')
axs[0].set_title('Top 10 Artists on YouTube')
axs[0].set_xlabel('Artist')
axs[0].set_ylabel('Total Views')
axs[0].tick_params(axis='x', rotation=90)

# Add the logo to the upper right corner of the chart
add_logo(axs[0], 'youtube_logo.png')

# Create a bar chart for top 10 artists on Spotify
axs[1].bar(grouped_data_sp.index, grouped_data_sp['Stream'], color='green')
axs[1].set_title('Top 10 Artists on Spotify')
axs[1].set_xlabel('\n\n\n\n\nArtist')
axs[1].set_ylabel('Total Streams')
axs[1].tick_params(axis='x', rotation=90)
axs[1].set_facecolor('black')  # Set the background color to black


# Add the logo to the upper right corner of the chart
add_logo(axs[1], 'spotify_logo.png')

# Adjust the layout of the subplots
plt.subplots_adjust(wspace=0.3)

# Show the plot
plt.show()

print('''Remarks:

    With 15 billion YouTube views, Ed Sheeran was the artist with the most viewers. With 14 billion views each for CoComelon and Katy Perry, they are the second and third most viewed channels in YouTube, respectively. The final two artists in the top five are Charlie Puth and Luis Fonsi, with 12 billion and 11 billion views, respectively.

    While using Spotify, Post Malone received 15 billion streams, followed by Ed Sheeran, who received 14 billion. With 13.5 billion and 13 billion streams each, Dua Lipa and XXXTENTACION are the third and fourth most streamed acts, respectively. The Weeknd has 13 billion streams and is in sixth position.

    Only Ed Sheeran and the Coldplay were able to make it to the Top 10 spots for most views/streams in both music streaming platform.

    These numbers demonstrate how well-liked these musicians and their music are across all of the numerous online channels.''')
```

```python
top_songs = df.sort_values('Stream', ascending=False).head(10)
top_songs[['Track', 'Energy', 'Danceability', 'Acousticness']]
```

```python
top_songs_melt = top_songs.melt(id_vars=['Track'], value_vars=['Energy', 'Danceability', 'Acousticness'],
                                var_name='Attribute', value_name='Value')

sns.set_style('whitegrid')
sns.catplot(x='Track', y='Value', hue='Attribute', data=top_songs_melt, kind='bar',
            palette={'Energy': 'green', 'Danceability': 'purple', 'Acousticness': 'yellow'},
            aspect=2, legend=False)


plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1))
plt.title('Top 10 Songs by Stream')
plt.xlabel('Track')
plt.ylabel('Value')

plt.xticks(rotation=75)
plt.show()

print('''Remarks:

    The graphical representation illustrates the music preference of Spotify users. The data reveals that the majority of
listeners favor songs with danceable qualities. Specifically, the energy level of almost every track on the platform is 0.5 or 
higher, indicating a strong preference for music that evokes feelings of excitement and vigor. In addition, roughly 50% of 
tracks on Spotify have an acousticness level above 0.5, which suggests a division among users as to whether they prefer more 
relaxing music. In conclusion, the findings demonstrate that music with dance beats and high energy tends to be more popular 
among Spotify users.
''')
```

```python
#Making the column "Artist" as the index
df_artist = df.set_index("Artist")
df_artist = df_artist[["Danceability","Views","Likes","Album_type"]]

# Group the data by artist and calculate the mean danceability score
df_artist_grouped = df_artist.groupby('Artist')['Likes'].sum().reset_index()
# Sort the dataframe by danceability score in descending order
df_artist_sorted = df_artist_grouped.sort_values(by='Likes', ascending=False)
# Select the top 10 artists with the highest danceability scores
top_10_artists = df_artist_sorted.head(10)
# Create a horizontal bar plot using seaborn
sns.barplot(x='Likes', y='Artist', data=top_10_artists)
# Set the axis labels and title
plt.xlabel('Number of Likes')
plt.ylabel('Artist')
plt.title('Number of Likes per Artist')
# Show the plot
plt.show()

print('''Remarks:
    
    According to the graph, the Korean sensational groups BTS and Blackpink garnered the highest number of likes per artist,
indicating their significant popularity among music listeners. This observation points to the remarkable success of K-pop
music, which has surpassed the long-standing dominance of western music in terms of audience preference.
''')
```

```python
# filter the data to select only singles
singles = df[df["Album_type"] == "single"]
# group the data by artist and calculate the total views and likes
grouped = singles.groupby("Artist").agg({"Views": "sum", "Likes": "sum"})
# sort the data based on likes and select the top 10 artists
top_10_artists = grouped.sort_values("Likes", ascending=False).head(10)
print(top_10_artists)

print('''

Remarks:

    The presented table presents the top 10 artists based on the views and likes of their single releases. With an astounding
7.3 billion views and 82 million likes, Blackpink secured the highest rank, confirming their preeminence in the music industry. 
Following closely behind with 6.7 billion views is Wiz Khalifa, while Charlie Puth and Ozuna hold the third and fourth 
positions with over 5 billion views each. Jubin Nautiyal completes the top five with 4.5 billion views. The sixth and seventh 
spots belong to Tanishk Bagchi and Lil Nas X, with over 3 billion views each. Marshmello and Billie Eilish occupy the eighth and ninth spots with over 2 billion views, respectively. The top 10 is completed by Bizarrap with 1.7 billion views.

    Overall, the dominance of Blackpink, the K-pop group, in both views and likes demonstrates the continuing global success of 
K-pop music in the music industry.
''')
```

```python
# filter the data to select only compilation
singles = df[df["Album_type"] == "compilation"]
# group the data by artist and calculate the total views and likes
grouped = singles.groupby("Artist").agg({"Views": "sum", "Likes": "sum"})
# sort the data based on likes and select the top 10 artists
top_10_artists = grouped.sort_values("Likes", ascending=False).head(10)
print(top_10_artists)

print('''

Remarks:
    
    The presented table presents the top 10 artists based on the views and likes of their compilation releases. With an 
astounding 3.4 billion views and 20 million likes, Shakira secured the highest rank, confirming her preeminence in the music 
industry. Following closely behind with 2.2 billion views is Ellie Goulding, while each of the remaining artist has over 1 
billion views, except for Annie Lennox with 863 million views.
''')
```
