```python
import requests

url = 'https://raw.githubusercontent.com/AllenDowney/ThinkPython2/master/code/words.txt'
response = requests.get(url)

with open('words.txt', 'wb') as f:
    f.write(response.content)

with open('words.txt', 'r') as f:
    contents = f.read()

# A. TOTAL NUMBER OF WORDS
word_list = contents.split()
num_words = len(word_list)
print("1) There a total of", num_words, "words in the file.")

# B. LONGEST WORD
longest_word = ""
for word in word_list:
    if len(word) > len(longest_word):
        longest_word = word

print("2) The longest word in the file is", longest_word + ".")

# C. WORDS STARTING WITH (USER INPUT)
while True:
    word_starting_letter = input("3) Enter the letter of the first word: ").lower()
    if len(word_starting_letter) == 1 and word_starting_letter.isalpha():
        break
    print("Invalid input. Please enter ONE LETTER only.")

word_start = []
for word in word_list:
    if word[0] == word_starting_letter:
        word_start.append(word)

while True:
    display_start = input("Display in list[] or not? [Y/N]: ").lower()
    if len(display_start) == 1 and display_start.isalpha():
        break
    print("Invalid input. Please enter ONE LETTER only.")

if display_start == 'Y' or display_start == 'y':
    print("Words starting with", word_starting_letter + ":", word_start)
else:
    print("Words starting with", word_starting_letter + ":")
    for word in word_start:
        print(word)

# D. WORDS CONTAINING A (USER INPUT)
print("")
while True:
    word_containing_letter = input("4) Enter the letter that you want to find in a word: ").lower()
    if len(word_containing_letter) == 1 and word_containing_letter.isalpha():
        break
    print("Invalid input. Please enter ONE LETTER only.")

word_contain = []
for word in word_list:
    for letter in word:
        if letter == word_containing_letter:
            word_contain.append(word)

while True:
    display_contain = input("Display in list[] or not? [Y/N]: ").lower()
    if len(display_contain) == 1 and display_contain.isalpha():
        break
    print("Invalid input. Please enter ONE LETTER only.")

if display_contain == 'Y' or display_contain == 'y':
    print("Words containing", word_containing_letter + ":", word_contain)
else:
    print("Words containing", word_containing_letter + ":")
    for word in word_contain:
        print(word)

# E. PALINDROMES
palindromes = []
for word in word_list:
    if word == word[::-1]:
        palindromes.append(word)

while True:
    display_palindromes = input("\nDisplay palindromes in list[] or not? [Y/N]: ").lower()
    if len(display_palindromes) == 1 and display_palindromes.isalpha():
        break
    print("Invalid input. Please enter ONE LETTER only.")

if display_palindromes == 'Y' or display_palindromes == 'y':
    print("5) The words that are palindromes in the file are:", palindromes)
else:
    print("5) The words that are palindromes in the file are:")
    for word in palindromes:
        print(word)

# F. LETTER FREQUENCY
letter_counts = {}
for word in word_list:
    for letter in word:
        if letter in letter_counts:
            letter_counts[letter] += 1
        else:
            letter_counts[letter] = 1

print("\n6) Frequency of each letter in every word in the file:")
for letter in sorted(letter_counts.keys()):
    print(letter, "=", letter_counts[letter])
```
