```python
import random


def game():
    # RANDOM WORD
    def random_word(word_list):
        return random.choice(word_list)

    # WORD UNDERSCORE
    def letter_string(random_word):
        letter_string = list(random_word)
        i = 0
        for letter in letter_string:
            letter_string[i] = "_"
            i += 1
        return letter_string

    # TAKE THE USER'S GUESSED LETTER
    def word_guess(revised_guessed_letter, guesses_list):
        if len(revised_guessed_letter) != 1:
            print("Invalid input. Please enter ONE LETTER only.\n")
            return False
        elif not revised_guessed_letter.isalpha():
            print("Invalid input. Please enter a LETTER only.\n")
            return False
        elif revised_guessed_letter in guesses_list:
            print("The letter has already been entered.\n")
            return False
        else:
            guesses_list.append(revised_guessed_letter)
            return True

    with open('words.txt', 'r') as f:
        contents = f.read()
        word_list = contents.split()

    print("Welcome to GUESS THE WORD!")
    while True:
        chosen_word = random_word(word_list)
        word_string = letter_string(chosen_word)
        number_of_guesses = 8
        guesses_list = []
        print("")
        while True:
            print(" ".join(word_string))

            guessed_letter = input("Guess a letter: ").lower()
            if not word_guess(guessed_letter, guesses_list):
                continue

            if guessed_letter in chosen_word:
                print("Correct!\n")
                for i in range(len(chosen_word)):
                    if guessed_letter == chosen_word[i]:
                        word_string[i] = guessed_letter
            else:
                number_of_guesses -= 1
                print("Incorrect letter! {} guesses remaining.\n".format(number_of_guesses))

            print("Used letters:", ", ".join(guesses_list))

            if "_" not in word_string:
                print(" ".join(word_string))
                print("CONGRATULATIONS! You've guessed the word.")
                break
            if number_of_guesses == 0:
                print("Sorry, you lose. The word was {}.".format(chosen_word))
                break

        while True:
            try_again = input("\nTry again? [Y/N]: ").lower()
            if try_again == 'n':
                break
            elif try_again == 'y':
                break
            else:
                print("Invalid input. Please enter 'Y' or 'N' only.")

        if try_again == 'n':
            break


game()
print("\nThank you for playing!")
```
