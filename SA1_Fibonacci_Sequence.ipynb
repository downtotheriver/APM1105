```python
import time

while True:
    # USING ITERATION
    def iteration(number_place):
        i = 0
        sequence = [1, 1]
        while i+2 < number_place:
            current_number = sequence[i] + sequence[i+1]
            sequence.append(current_number)
            i += 1
        return sequence[number_place-1]


    # USING RECURSION
    def recursion(number_place, initializer, i, sequence):
        if initializer < number_place:
            current_number = sequence[i] + sequence[i+1]
            sequence.append(current_number)
            recursion(number_place, initializer+1, i+1, sequence)
        return sequence[number_place-1]


    while True:
        try:
            number_place = int(input("Enter the nth number: "))
            if number_place > 0:
                break
            else:
                print("Number must be positive.")
        except ValueError:
            print("Invalid input. Please enter a positive integer.")


    start_time = time.perf_counter()
    result_1 = iteration(number_place)
    end_time = time.perf_counter()
    iteration_runtime = end_time - start_time

    start_time = time.perf_counter()
    result_2 = recursion(number_place, initializer=2, i=0, sequence=[1, 1])
    end_time = time.perf_counter()
    recursive_runtime = end_time - start_time

    print("Iteration result:", result_1)
    print("Iterative rending time:", iteration_runtime, "seconds")
    print("Recursive result:", result_2)
    print("Recursive rending time:", iteration_runtime, "seconds")

    if iteration_runtime < recursive_runtime:
        print("Iteration is faster.")
    else:
        print("Recursion is faster.")

    while True:
        try_again = input("\nTry again? [Y/N]: ").lower()
        if try_again == 'n':
            break
        elif try_again == 'y':
            break
        else:
            print("Invalid input. Please enter 'Y' or 'N' only.")

    if try_again == 'n':
        break

print("Thank you!")
```
